# About Dataset
The dataset focuses on the health and lifestyle of individuals, aiming to explore factors contributing to obesity. It includes features such as gender, age, height, weight, family history of overweight, dietary habits, physical activity, and transportation mode. Each entry represents an individual, providing a diverse set of information for analyzing correlations between lifestyle choices and obesity. The dataset offers insights into the complex interplay of factors influencing health outcomes, making it valuable for investigating patterns in the relationship between various aspects of individuals' lives and their likelihood of being overweight or obese.
### Data Dictionary
1. **Gender**: The gender of the individual (e.g., Female, Male).
2. **Age**: The age of the individual.
3. **Height**: The height of the individual.
4. **Weight**: The weight of the individual.
5. **Family_history_with_overweight**: Whether there is a family history of overweight (yes/no).
6. **FAVC**: Frequency of high caloric food consumption (yes/no).
7. **FCVC**: Frequency of consumption of vegetables.
8. **NCP**: Number of main meals.
9. **CAEC**: Consumption of food between meals.
10. **SMOKE**: Smoking habit (yes/no).
11. **CH2O**: Daily water consumption.
12. **SCC**: Calories consumption monitoring (yes/no).
13. **FAF**: Physical activity frequency.
14. **TUE**: Time using technology devices.
15. **CALC**: Consumption of alcohol.
16. **MTRANS**: Mode of transportation.
17. **NObeyesdad**: Obesity status.

```python
# Import all the required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
# Import Label encoder for encoding the data values
label_encoder = preprocessing.LabelEncoder()
```

```python
# Import Dataset
df = pd.read_csv('ObesityDataSet.csv')
```

## 1. Descriptive Statistics:

1. What is the distribution of age, height, and weight in the dataset?
2. How many males and females are in the dataset?

4. What is the most common mode of transportation (MTRANS) in the dataset?

```python
## What is the distribution of age, height, and weight in the dataset?
# Set up a 1x3 grid for subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Plot the distribution of Age
sns.histplot(df['Age'], bins=20, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Age')

# Plot the distribution of Height
sns.histplot(df['Height'], bins=20, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Height')

# Plot the distribution of Weight
sns.histplot(df['Weight'], bins=20, kde=True, ax=axes[2])
axes[2].set_title('Distribution of Weight')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# How many Males and Females are in the dataset?
gender_colors = {'Male': 'blue', 'Female': 'pink'}

ax = sns.countplot(data=df, x='Gender', hue='Gender', palette=gender_colors)
for label in ax.containers:
    ax.bar_label(label)
```

```python
# The average daily water consumption
average_water_consumption = df['CH2O'].mean()
average_water_consumption
```

## 5. **Transportation and Lifestyle:**

#### Does the mode of transportation (MTRANS) have any impact on obesity status?

```python

plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='MTRANS', hue='NObeyesdad', palette='Set1')
for label in ax.containers:
    ax.bar_label(label, fmt='%d')

plt.title('Impact of Mode of Transportation on Obesity Status')
plt.xlabel('Mode of Transportation (MTRANS)')
plt.ylabel('Count')
plt.legend(title='Obesity Status')
plt.xticks(ha='right')
plt.show()
```

#### - Are there differences in obesity patterns based on the mode of transportation for males and females?

```python
# Count plot of MTRANS by Obesity Status for Males
plt.figure(figsize=(14, 7))
ax = sns.countplot(data=df[df['Gender']=='Male'], x='MTRANS', hue='NObeyesdad', palette='Set1', hue_order=df['NObeyesdad'].value_counts().index)
for label in ax.containers:
    ax.bar_label(label, fmt='%d')

plt.title('Obesity Patterns based on Mode of Transportation for Males')
plt.xlabel('Mode of Transportation (MTRANS)')
plt.ylabel('Count')
plt.legend(title='Obesity Status')
plt.xticks(rotation=45, ha='right')
plt.show()

# Count plot of MTRANS by Obesity Status for Females
plt.figure(figsize=(14, 7))
ax = sns.countplot(data=df[df['Gender']=='Female'], x='MTRANS', hue='NObeyesdad', palette='Set1', hue_order=df['NObeyesdad'].value_counts().index)
for label in ax.containers:
    ax.bar_label(label, fmt='%d')

plt.title('Obesity Patterns based on Mode of Transportation for Females')
plt.xlabel('Mode of Transportation (MTRANS)')
plt.ylabel('Count')
plt.legend(title='Obesity Status')
plt.xticks(rotation=45, ha='right')
plt.show()
```
